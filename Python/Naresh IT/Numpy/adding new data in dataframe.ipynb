```python
import pandas as pd
```

```python
df = pd.read_csv("./Student_details.csv")
df
```

```python
df["total"] = df["Phy"]+df["Che"]+df["Math"]+df["IT"]
```

```python
df
```

```python
df["Percentage"] = df["total"]*100/400
df
```

```python
df["Percentage"].describe()
```

```python
df.loc[df['Percentage']>70]
```

```python
df["Grade"] = "null"
df
```

```python
df.loc[(df["Percentage"]>=80),["Grade"]] = "A"
df
```

```python
df.loc[(df["Percentage"]<80.00) & (df["Percentage"]>=60.00),["Grade"]] = "B"
df
```

```python
df.loc[(df["Percentage"]<60.00) & (df["Percentage"]>=40.00),["Grade"]] = "C"
df
```

```python
df.loc[(df["Percentage"]<40.00),["Grade"]] = "Failled"
df
```
